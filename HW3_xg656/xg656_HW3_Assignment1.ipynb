```python
def plotdistributionpoint(dis_type):
    mean_collect=[]
    for i in size:
        var_normal=getattr(np.random, dis_type)(df, size = i)
        mean_normal=var_normal.mean()
        mean_collect.append(mean_normal)
        plot(i,mean_normal,'-o')
        plt.xlabel('sample size')
        plt.ylabel('sample mean')
        plt.title(dis_type+' distribution')
    plot([min(size), max(size)], [df, df], 'k')
    return mean_collect
```

```python
def normfit(mean_get):
    from scipy.stats import norm
    param = norm.fit(mean_get) # distribution fitting
    x = np.linspace(95,105,100)
    fit = norm.pdf(x,loc=param[0],scale=param[1])
    plt.title('Normal fit of previous distribution')
    plt.plot(x,fit)
    plt.hist(mean_get,normed=1)
    plt.show()
    return
```

```python
mean_get=plotdistributionpoint('poisson')
```

```python
normfit(mean_get)
```

```python
mean_get=plotdistributionpoint('chisquare')
```

```python
normfit(mean_get)
```

```python
   
mean_collect=[]
for i in size:
    var_binomial=np.random.binomial(200, 0.5,i)
    mean_binomial=var_binomial.mean()
    mean_collect.append(mean_binomial)
    plot(i,mean_binomial,'-o')
    plt.xlabel('sample size')
    plt.ylabel('sample mean')
    plt.title('binomial'+' distribution')
plot([min(size), max(size)], [df, df], 'k')
```

```python
normfit(mean_get)
```
