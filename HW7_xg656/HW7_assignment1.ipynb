```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```


# Starbucks has more stores in Manhattan than any other national retailer, with 212 locations. In each of the other boroughs, Dunkin Donuts tops the listâ€”it has 154 stores in Queens, 123 in Brooklyn, 72 in the Bronx and 32 on Staten Island.

```python
Y=[212,154,123,72,32]
X=[0,1,2,3,4]
X1=[0.25,1.25,2.25,3.25,4.25]
width=0.5
xticks = ['Manhattan','Queens','Brooklyn','Bronx','Staten Island']
```

```python
import urllib2
ax=plt.bar(X,Y,width,color="#0000FF",alpha=0.5)
b=plt.xticks(X1, xticks,fontsize=12)

ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["right"].set_visible(False)  

ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
for i in range(0, 250, 50):    
    plt.plot(range(-1,6), [i] * len(range(-1,6)), "--", lw=1, color="black", alpha=0.3) 
    
plt.xlim(-0.25,4.75)
plt.ylabel('Number of Starbuck stores',fontsize=12)
plt.title('Number of Starbuck stores in five boroughs of NYC (2016)\n',fontsize=12)
plt.show()
```
